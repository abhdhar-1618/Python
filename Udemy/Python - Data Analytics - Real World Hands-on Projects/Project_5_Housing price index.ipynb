```python
#df1.groupby('year').average_price.mean().round(2)

mean_values=df1.groupby('year').average_price.mean().round(2)

# Plot the mean values using pandas plot method
mean_values.plot(kind='line', marker='o', linestyle='-', color='r', figsize=(10, 6))

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Mean Average Price')
plt.title('Mean Average Price Over the Years')

# Display the plot
plt.show()
```

```python
from datetime import datetime

# Read the external file into a DataFrame
file_path = (r"D:\Main\My Practice and experiments\My Practice LAB\Python\2023\Udemy\Python - Data Analytics - Real World Hands-on Projects\Project_5\HPI.csv")  # Replace with the actual file path
df = pd.read_csv(file_path)

# Assuming your date column is named 'date_column', replace it with your actual column name
date_column = df['date']

# Iterate through each row and extract day and month
for datestamp in date_column:
    date_object = datetime.strptime(datestamp, "%m/%d/%Y")
    day = date_object.day
    month = date_object.month
    year = date_object.year
    # Do something with day and month, for example, print them
    print(f"Day: {day}, Month: {month}, Year: {year}")
```
