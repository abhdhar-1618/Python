```python
from itertools import permutations

def find_anagrams(words):
    all_anagrams = set()
    for perm in permutations(words):
        anagram = ''.join(perm)
        all_anagrams.add(anagram)
    return all_anagrams

def main():
    input_words = input("Enter words separated by spaces: ").split()
    all_combinations = []
    for i in range(1, len(input_words) + 1):
        all_combinations.extend(permutations(input_words, i))

    for combination in all_combinations:
        anagrams = find_anagrams(combination)
        print(f"Anagrams of {' '.join(combination)}:")
        for anagram in anagrams:
            print(anagram)
        print()

if __name__ == "__main__":
    main()
```

```python
from itertools import combinations, permutations

def generate_anagrams(word):
    anagrams = set()
    max_word_length = min(len(word), 10)  # Limit the maximum length of word to consider
    for i in range(1, max_word_length + 1):
        for combo in combinations(word, i):
            if len(combo) > 7:  # Limit the number of characters in each combination
                continue
            for perm in permutations(combo):
                anagram = ''.join(perm)
                anagrams.add(anagram)
                if len(anagrams) >= 100:  # Limit the total number of anagrams
                    return anagrams
    return anagrams

def main():
    original_word = input("Enter the original word or phrase: ")
    target_word = input("Enter the target word or phrase you want to transform it into: ")

    choice = input("Do you want to (A)dd, (R)emove, or (K)eep the letters as is? ").lower()
    
    original_anagrams = generate_anagrams(original_word)

    if choice == 'a':
        target_anagrams = set()
        for anagram in original_anagrams:
            for i in range(len(anagram) + 1):
                for char in target_word:
                    new_anagram = anagram[:i] + char + anagram[i:]
                    target_anagrams.add(new_anagram)
                    if len(target_anagrams) >= 100:
                        break
                if len(target_anagrams) >= 100:
                    break
            if len(target_anagrams) >= 100:
                break
        
        print("Anagrams:")
        print("\n".join(list(target_anagrams)[:100]))  # Limit the number of displayed anagrams
    
    elif choice == 'r':
        target_anagrams = set()
        for anagram in original_anagrams:
            for i in range(len(anagram)):
                new_anagram = anagram[:i] + anagram[i+1:]
                target_anagrams.add(new_anagram)
                if len(target_anagrams) >= 100:
                    break
            if len(target_anagrams) >= 100:
                break
        
        print("Anagrams:")
        print("\n".join(list(target_anagrams)[:100]))  # Limit the number of displayed anagrams
    
    elif choice == 'k':
        print("Anagrams:")
        print("\n".join(list(original_anagrams)[:100]))  # Limit the number of displayed anagrams
    
    else:
        print("Invalid choice!")

if __name__ == "__main__":
    main()
```
