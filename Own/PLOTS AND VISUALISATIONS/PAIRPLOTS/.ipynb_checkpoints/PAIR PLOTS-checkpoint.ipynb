```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd

# Fetch dataset 
solar_flare = fetch_ucirepo(id=89) 

# Data (as pandas DataFrames) 
X = solar_flare.data.features 
y = solar_flare.data.targets 

# Metadata 
metadata = solar_flare.metadata 
variables = solar_flare.variables 

# Print metadata in a readable format
print("Metadata:")
print("===================")
for key, value in metadata.items():
    print(f"{key}: {value}")

# Print variable information in a readable format
print("\nVariable Information:")
print("===================")
for variable in variables:
    print(f"Variable: {variable}")
    print("-------------------")
```

# seaborn.PairGrid

```python
f = sns.PairGrid(flares)
```

```python
f = sns.PairGrid(flares)
f.map(sns.scatterplot)
```

```python
f = sns.PairGrid(flares)
f.map_diag(sns.histplot)
f.map_offdiag(sns.scatterplot)
```

```python
f = sns.PairGrid(flares, diag_sharey=False)
f.map_upper(sns.scatterplot)
f.map_lower(sns.kdeplot)
f.map_diag(sns.kdeplot)
```

```python
f = sns.PairGrid(flares, diag_sharey=False, corner=True)
f.map_lower(sns.scatterplot)
f.map_diag(sns.kdeplot)
```

```python
f = sns.PairGrid(flares, hue="modified Zurich class")
f.map_diag(sns.histplot)
f.map_offdiag(sns.scatterplot)
f.add_legend()
```

```python
f = sns.PairGrid(flares, hue="spot distribution")
f.map_diag(plt.hist)
f.map_offdiag(plt.scatter)
f.add_legend()
```

```python
f = sns.PairGrid(flares, hue="modified Zurich class")
f.map_diag(sns.histplot)
f.map_offdiag(sns.scatterplot, size=flares["area"])
f.add_legend(title="Zurich Class", adjust_subtitles=True)
```

```python
f = sns.PairGrid(flares, hue="spot distribution")
f.map_diag(sns.histplot, hue=None, color=".3")
f.map_offdiag(sns.scatterplot)
f.add_legend()
```

```python
variables = ["common flares", "moderate flares", "severe flares"]
f = sns.PairGrid(flares, hue="area of largest spot", vars=variables)
f.map_diag(sns.histplot, hue=None, color=(0, 1, 0))
f.map_offdiag(sns.scatterplot)
f.add_legend()
```

```python
x_vars = ["common flares", "moderate flares", "severe flares"]
y_vars = ["area of largest spot"]
g = sns.PairGrid(flares, hue="area of largest spot", x_vars=x_vars, y_vars=y_vars)
g.map_diag(sns.histplot, color=".3")
g.map_offdiag(sns.scatterplot)
g.add_legend()
```

```python
f = sns.PairGrid(flares, hue="evolution")
f.map_diag(sns.histplot, multiple="stack", element="step")
f.map_offdiag(sns.scatterplot)
f.add_legend()
```
