## All in one script

```python
import pandas as pd

# Read the CSV file into df dataframe
df = pd.read_csv('emissions.csv')

# Remove specified columns
columns_to_remove = ['Series Name', 'Series Code']
df.drop(columns=columns_to_remove, inplace=True)

# Rename columns
columns_to_rename = df.columns[df.columns.str.contains('\[\w+\]')]
new_column_names = [col.split(' ')[0] for col in columns_to_rename]
rename_dict = dict(zip(columns_to_rename, new_column_names))
df.rename(columns=rename_dict, inplace=True)

# Replace '..' values with NaN
numerical_columns = df.select_dtypes(include='number').columns
df[numerical_columns] = df[numerical_columns].replace('..', 0 )

# Convert numerical columns to numeric data type
df[numerical_columns] = df[numerical_columns].apply(pd.to_numeric)

# Round numerical columns to two decimal points
# df[numerical_columns] = df[numerical_columns].round(2)

# Remove last 5 rows
df = df.iloc[:-5]

# Melt the DataFrame
data = df.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Value')
# data['Value'] = data['Value'].round(2)
# round(data.Value(), 2)

data['Value'] = pd.to_numeric(data['Value'], errors='coerce')  # Convert to numeric, coercing errors
data['Value'] = data['Value'].round(2)  # Round off to two decimal places

# Save the formatted data into a new dataframe named "data"
data.to_csv('emissions_modified.csv', index=False)
```

## Merging two files 

```python
import pandas as pd

# Read the modified CSV files
electricity_df = pd.read_csv('electricity_access_modified.csv')
agricultural_df = pd.read_csv('agricultural_land_modified.csv')

# Merge the two DataFrames based on "Country Code" and "Year"
merged_df = pd.merge(electricity_df, agricultural_df, on=['Country Code', 'Year'], how='left')

# Save the merged DataFrame to a new CSV file
merged_df.to_csv('electricity_&_agriculture.csv', index=False)
```

## Picking all files from a single location to merge

```python
import pandas as pd
import os

# Directory containing the CSV files
directory = r'D:\Main\My Practice and experiments\My Practice LAB\Python\2024\Own Projects\Data_Transformation\Converted'

# List to store DataFrames
dataframes = []

# Iterate over files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        filepath = os.path.join(directory, filename)
        df = pd.read_csv(filepath)
        dataframes.append(df)

# Merge the DataFrames based on "Country Code" and "Year"
merged_df = dataframes[0]  # Start with the first DataFrame
for df, filename in zip(dataframes[1:], os.listdir(directory)[1:]):
    # Extract the filename without extension
    filename = os.path.splitext(filename)[0]
    # Rename columns in the current DataFrame with filename suffixes
    df.columns = [f'{col}_{filename}' if col not in ['Country Code', 'Year'] else col for col in df.columns]
    # Merge with the main DataFrame
    merged_df = pd.merge(merged_df, df, on=['Country Code', 'Year'], how='left')

# Save the merged DataFrame to a new CSV file
merged_df.to_csv('merged_data.csv', index=False)
```
