```python
from docx import Document

def find_words_around_specific_word(docx_file_path, target_word, num_words_before=3, num_words_after=3):
    # Load the Word document from the file path
    doc = Document(docx_file_path)
    
    # Initialize an empty list to store the results
    results = []
    
    # Iterate through paragraphs in the document
    for paragraph in doc.paragraphs:
        # Split paragraph text into words
        words = paragraph.text.split()
        
        # Iterate through words in the paragraph
        for i, word in enumerate(words):
            # Check if the current word matches the target word
            if word.lower() == target_word.lower():
                # Get the words before and after the target word
                words_before = words[max(0, i - num_words_before) : i]
                words_after = words[i + 1 : i + num_words_after + 1]
                
                # Append the results to the list
                results.append((words_before, target_word, words_after))
    
    return results

if __name__ == "__main__":
    docx_file_path = r"D:\Main\My Practice and experiments\My Practice LAB\Python\2024\NLP\wordfind.docx"  # Replace with the actual file path
    target_word = "analyst"    # Word to search for
    num_words_before = 3       # Number of words before the target word
    num_words_after = 0        # Number of words after the target word
    
    results = find_words_around_specific_word(docx_file_path, target_word, num_words_before, num_words_after)
    
    # Print the results
    for words_before, target_word, words_after in results:
        print("Words before:", " ".join(words_before))
        print("Target word:", target_word)
        print("Words after:", " ".join(words_after))
        print()
```
