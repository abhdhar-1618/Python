# Data Preprocessing Tools

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import shapiro
import seaborn as sns
```

## Importing the dataset

```python
dataset = pd.read_csv('synthetic_model_ready_data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

# Dataset overview

```python
dataset.describe()
```

# Data normality check

```python
normality_results = []

# Iterate over each column in the DataFrame
for column in dataset.columns:
    # Attempt to convert the data in the column to numeric type
    try:
        numeric_data = pd.to_numeric(dataset[column])
        # Perform Shapiro-Wilk test on the numeric data
        stat, p = shapiro(numeric_data)
        
        # Check if the p-value is less than 0.05 (common significance level)
        if p > 0.05:
            result = "Normally Distributed"
        else:
            result = "Not Normally Distributed"
        
        # Append the result to the list
        normality_results.append((column, result))
    except ValueError:
        # Skip columns that cannot be converted to numeric type
        continue

# Convert the list of results into a DataFrame for easy viewing
normality_df = pd.DataFrame(normality_results, columns=["Column", "Normality"])
print(normality_df)
```

# Statistical visualization

```python
# Iterate over each column in the dataset
for column in dataset.columns:
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Plot a histogram for the column
    sns.histplot(data=dataset[column], kde=True, ax=ax, color='skyblue', bins=30)
    
    # Alternatively, plot a KDE plot
    # sns.kdeplot(data=dataset[column], ax=ax, color='skyblue')
    
    # Set title and labels
    ax.set_title(f'Distribution of {column}')
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')
    
    # Show plot
    plt.show()
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the Random Forest Classification model on the Training set

```python
# training method_1
'''from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)'''
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for grid search
param_grid = {
    'n_estimators': [10, 50, 100, 200, 400],  # Try different numbers of trees
    'max_depth': [None, 10, 20],      # Try different maximum depths of trees
    'min_samples_split': [2, 5, 10],  # Try different minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4]     # Try different minimum number of samples required at each leaf node
}

# Create a RandomForestClassifier
classifier = RandomForestClassifier(random_state=0)

# Perform grid search with cross-validation
grid_search = GridSearchCV(estimator=classifier, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Get the best parameters found by grid search
best_params = grid_search.best_params_
print("Best parameters:", best_params)

# Train a RandomForestClassifier with the best parameters
best_classifier = RandomForestClassifier(**best_params, random_state=0)
best_classifier.fit(X_train, y_train)

# Evaluate the model on the test set
accuracy = best_classifier.score(X_test, y_test)
print("Accuracy:", accuracy)
```
