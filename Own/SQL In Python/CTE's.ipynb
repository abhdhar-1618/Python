# Question 1:How do you select all addresses from the address table using a CTE?

```python
%%sql
WITH AddressCTE AS (
    SELECT * FROM address
)
SELECT * FROM AddressCTE;
```

# Question 2:How can you count the number of addresses in the address table using a CTE?

```python
%%sql

WITH AddressCount AS (
    SELECT COUNT(*) AS TotalAddresses FROM address
)
SELECT * FROM AddressCount;
```

# Question 3:How do you find the maximum address_id using a CTE?

```python
%%sql

WITH MaxAddressID AS (
    SELECT MAX(address_id) AS MaxID FROM address
)
SELECT * FROM MaxAddressID;
```

# Question 4:How can you list all cities with their respective city_id using a CTE?

```python
%%sql

WITH CityList AS (
    SELECT city_id, city FROM city
)
SELECT * FROM CityList;
```

# Question 5:How do you get all unique district values from the address table using a CTE?

```python
%%sql

WITH Districts AS (
    SELECT DISTINCT district FROM address
)
SELECT * FROM Districts;
```

# Question 6:How do you list all addresses along with their city names using a CTE to join address and city tables?

```python
%%sql

WITH AddressCity AS (
    SELECT a.address_id, a.address, c.city
    FROM address a
    JOIN city c ON a.city_id = c.city_id
)
SELECT * FROM AddressCity;
```

# Question 7:How can you find the average address_id grouped by district using a CTE?

```python
%%sql
WITH AverageAddressID AS (
    SELECT district, AVG(address_id) AS AvgID
    FROM address
    GROUP BY district
)
SELECT * FROM AverageAddressID;
```

# Question 8:How do you list all addresses with their corresponding city names where the district is 'New York'?

```python
%%sql
WITH AddressCity AS (
    SELECT a.address_id, a.address, c.city
    FROM address a
    JOIN city c ON a.city_id = c.city_id
    WHERE a.district = 'New York'
)
SELECT * FROM AddressCity;
```

# Question 9:How can you get the number of addresses in each district using a CTE?

```python
%%sql
WITH DistrictCount AS (
    SELECT district, COUNT(*) AS AddressCount
    FROM address
    GROUP BY district
)
SELECT * FROM DistrictCount;
```

# Question 10:How do you get the top 5 districts by number of addresses using a CTE?

```python
%%sql

WITH DistrictRanking AS (
    SELECT district, COUNT(*) AS AddressCount
    FROM address
    GROUP BY district
    ORDER BY AddressCount DESC
)
SELECT * FROM DistrictRanking
LIMIT 5;
```

# Question 11:How do you find the second highest address_id in the address table using a CTE?

```python
%%sql

WITH RankedAddresses AS (
    SELECT address_id, ROW_NUMBER() OVER (ORDER BY address_id DESC) AS rn
    FROM address
)
SELECT address_id
FROM RankedAddresses
WHERE rn = 2;
```

# Question 12:How can you get the average number of addresses per city where the city has more than 10 addresses?

```python
%%sql
WITH CityAddressCount AS (
    SELECT c.city_id, c.city, COUNT(a.address_id) AS AddressCount
    FROM city c
    JOIN address a ON c.city_id = a.city_id
    GROUP BY c.city_id, c.city
    HAVING COUNT(a.address_id) > 10
)
SELECT AVG(AddressCount) AS AvgAddresses
FROM CityAddressCount;
```

# Question 13:How do you list all districts with the total number of addresses and order them by the number of addresses in descending order?

```python
%%sql

WITH DistrictSummary AS (
    SELECT district, COUNT(*) AS AddressCount
    FROM address
    GROUP BY district
)
SELECT district, AddressCount
FROM DistrictSummary
ORDER BY AddressCount DESC;
```

# Question 14:How can you find the address with the maximum address_id in each district?

```python
%%sql

WITH MaxAddressPerDistrict AS (
    SELECT district, MAX(address_id) AS MaxID
    FROM address
    GROUP BY district
)
SELECT a.address_id, a.address, a.district
FROM address a
JOIN MaxAddressPerDistrict m ON a.district = m.district AND a.address_id = m.MaxID;
```

# Question 15:How do you find the number of unique addresses in each city, considering only cities with more than 5 unique addresses?

```python
%%sql
WITH CityAddressCount AS (
    SELECT c.city, COUNT(DISTINCT a.address_id) AS UniqueAddressCount
    FROM city c
    JOIN address a ON c.city_id = a.city_id
    GROUP BY c.city
    HAVING COUNT(DISTINCT a.address_id) > 5
)
SELECT * FROM CityAddressCount;
```
