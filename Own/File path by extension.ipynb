# File path by extension

```python
import os
import pandas as pd

def find_files_in_directory(directory_path, extensions):
    file_info_list = []

    for root, dirs, files in os.walk(directory_path):
        print("Current directory:", root)
        for filename in files:
            try:
                if filename.endswith(tuple(extensions)):
                    file_path = os.path.join(root, filename)
                    file_info_list.append({
                        'File': filename,
                        'File Path': file_path
                    })
            except Exception as e:
                print(f"Error processing file {filename}: {str(e)}")
                continue

    return file_info_list

if __name__ == "__main__":
    directory = 'D:\\'
    extensions_to_search = ['.sql']  # Add more extensions as needed
    # extensions_to_search = ['.ipynb', '.docx', '.txt']  # Add more extensions as needed

    file_info_list = find_files_in_directory(directory, extensions_to_search)
    
    # Create a DataFrame from the list of file information
    df = pd.DataFrame(file_info_list)

    # Specify the desired location to save the CSV file
    save_location = r"D:\Main\My Practice and experiments\My Practice LAB\Python\2023\Movie recommendation project\File Informations"  # Change this to your desired location

    # Export the DataFrame to a CSV file at the specified location
    csv_filename = "D_Drive_SQL_file_info_test.csv"
    csv_filepath = os.path.join(save_location, csv_filename)
    df.to_csv(csv_filepath, index=False)

    print(f"File information exported to {csv_filepath}")
    
    
# '''    # Create a DataFrame from the list of file information
#     df = pd.DataFrame(file_info_list)

#     # Export the DataFrame to a CSV file
#     csv_filename = "G_Drive_Video_file_info.csv"
#     df.to_csv(csv_filename, index=False)

#     print(f"File information exported to {csv_filename}")'''
```
