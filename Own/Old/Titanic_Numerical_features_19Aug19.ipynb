### Concatenate Train and Test & Check Missing Values

```python
titanic = pd.concat([train,test], sort = False)
titanic.describe()
```

### Check Missing Values using map

```python
titanic.info()
```

### Treat Age & Fare

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
titanic['Fare'] = titanic['Fare'].fillna(titanic['Fare'].median())
titanic.info()
```

## Model 1 : Var for Model

```python
titanic_all = titanic.drop(['PassengerId','Name','Cabin','Embarked', 'Ticket'],axis=1)
```

```python
titanic_all.head()
```

### Divide Dataset into train and test after cleaning missing values

```python
titanic_all['Sex'] = titanic_all['Sex'].map({'male': 1, 'female': 0})
titanic_all.info()
```

```python
#Get Train
t_train = titanic_all[titanic_all['Survived'].notnull()]
t_train.describe()
```

```python
#Get Test
s_test = titanic_all[titanic_all['Survived'].isnull()]
s_test.describe()
```

```python
t_train.head()
```

```python
X = t_train.iloc[:,1:]
X.head()
```

```python
y = t_train['Survived']
y.head()
```

```python
import sklearn.model_selection as model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y,test_size=0.2, random_state = 200)
```

### Log Reg

```python
#Import libraries
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression(solver = 'lbfgs')
```

```python
logreg.fit(X_train,y_train)
logreg.score(X_test,y_test)
```

```python
t_predictions = logreg.predict(X_test)
t_predictions
```

## Evaluation on test

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,t_predictions))
```

## Prediction for Submission test 

```python
t_predictions = logreg.predict(s_test.iloc[:,1:])
t_predictions
```
